```python
import pandas as pd
import matplotlib.pyplot as plt

print(pd.__version__)
```

```python
dataset = pd.read_csv('imdb_top_1000.csv')
```

```python
names = dataset.columns.tolist()
print(names)
```

```python
names_df = pd.DataFrame({'Column_names': names})
names_df.to_csv('column_names.csv', index = False)
```

```python
average_rating = dataset['IMDB_Rating'].mean()
average_rating_df = pd.DataFrame({'Average Rating': [average_rating]})
average_rating_df.to_csv('average_rating.csv', index = False)
```

```python
unique_genres = dataset['Genre'].nunique()
unique_genres_df = pd.DataFrame({'Unique Genres': [unique_genres]})
unique_genres_df.to_csv('unique_genres.csv', index = False)
```

```python
# Plotting the histogram
genres = dataset['Genre']
genre_count = genres.value_counts()
plt.figure(figsize=(20, 6))
plt.bar(genre_count.index, genre_count.values)
plt.xlabel('Genres')
plt.ylabel('Count')
plt.title('Genre Distribution of Movies')
plt.xticks(rotation='vertical', ha='right')
plt.subplots_adjust(bottom=0.2)
plt.tight_layout(pad=2.0)
plt.show()
```

```python
year = dataset['Released_Year']
movies_per_year = year.value_counts().sort_index()
years = movies_per_year.index
movie_counts = movies_per_year.values
plt.figure(figsize=(10, 6))
plt.scatter(years, movie_counts)
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.title('Number of Movies Made Each Year')
plt.grid(True)
plt.tight_layout()
plt.show()
```
